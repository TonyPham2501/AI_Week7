```python
import pandas as pd
url = '/content/Salary_Data.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt 
data.hist()
plt.show 
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns 
sns.heatmap(correlation,annot = True)
```

```python
data.plot(kind ='scatter',x = 'YearsExperience',y = 'Salary',figsize =(9,9),color ='black')
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['YearsExperience']), y = data['Salary'])
print(model.intercept_)
print(model.coef_)

#y = coef*x +_ intercerpt 
```

```python
print(model.score(X = pd.DataFrame(data['YearsExperience']), y = data['Salary']))
```

```python
data.plot(kind ='scatter',x= 'YearsExperience',y='Salary',figsize= (9,9), color = 'black', )
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2]).T
model.fit(X,y = data['Salary'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Salary']))
```

```python

import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3]).T
model.fit(X,y = data['Salary'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Salary']))
```

```python
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3,data['YearsExperience']**4,data['YearsExperience']**5]).T
model.fit(X,y = data['Salary'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Salary']))
```

```python
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```
