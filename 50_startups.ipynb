```python
data = pd.read_csv(url)
import seaborn as sns 
sns.pairplot(data)
```

```python
data.plot(kind ='scatter',x = 'R&D Spend',y = 'Profit',figsize =(9,9),color ='black')
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['R&D Spend']), y = data['Profit'])
print(model.intercept_)
print(model.coef_)

#y = coef*x +_ intercerpt 
```

```python
print(model.score(X = pd.DataFrame(data['R&D Spend']), y = data['Profit']))
```

```python
data.plot(kind ='scatter',x= 'R&D Spend',y='Profit',figsize= (9,9), color = 'black', )
y_pred = model.predict(X = pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2]).T
model.fit(X,y = data['Profit'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Profit']))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3]).T
model.fit(X,y = data['Profit'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Profit']))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3,data['R&D Spend']**4,data['R&D Spend']**5]).T
model.fit(X,y = data['Profit'])
print('Intercept cua mo hinh: ', model.intercept_)
print('Coeffient cua mo hinh: ', model.coef_)
print('Do chinh xac cua mo hinh: ', model.score(X,y = data['Profit']))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```
